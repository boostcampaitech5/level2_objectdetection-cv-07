#### Individual Image

```python
def individual_image(mode:str, id:str, config_file:str, checkpoint_file:str, score_threshold:int):
    model = init_detector(config_file, checkpoint_file, device='cuda:0')
    img = mmcv.imread(f'../../../dataset/{mode}/{id}.jpg')
    result = inference_detector(model, img)
    show_result_pyplot(model, img, result, score_thr=score_threshold, title=f'{mode}/{id}.jpg')
```

```python
# test dataset
individual_image(mode="test", id="4870", config_file='../woo/swin-cascadeRCNN2.py', checkpoint_file='../woo/epoch_26.pth', score_threshold=0.3)

# train dataset
#individual_image("train", "4882", '../woo/swin-cascadeRCNN2.py', '../woo/epoch_26.pth', 0.3)
```
